```python
data=np.array([10,12,13,15,18,20,105,120])
```

```python
mean = np.mean(data)
```

```python
mean
```

```python
std=np.std(data)
```

```python
std
```

```python
z_scores=(data-mean)/std
```

```python
z_scores
```

```python
np.abs(z_scores)>1
```

```python
outliers=data[np.abs(z_scores)>1]
```

```python

outliers
```
