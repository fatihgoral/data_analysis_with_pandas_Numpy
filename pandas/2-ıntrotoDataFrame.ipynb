```python
first_index = ["Argentina","Argentina","Portugal",
               "Brazil","Brazil","France"]
```

```python
inner_index = ["Messi","Di Maria","Ronaldo",
               "Neymar","Ronaldinho","Mbappe"]
```

```python
zipped_index = list(zip(first_index, inner_index))
```

```python
zipped_index
```

```python
zipped_index=pd.MultiIndex.from_tuples(zipped_index)
```

```python
zipped_index
```

```python
big_df=pd.DataFrame(sample_values,index=zipped_index,columns=["Age","Goals"])
```
